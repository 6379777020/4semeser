```python
tip = ng.load_dataset('tips')
```

```python
tip
```

```python
tip.tail()
```

```python
tip.head()
```

# relationl plot

```python
ng.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day",style="day",kind="scatter")
```

```python
ng.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day",style="day",kind="line")
```

# box plot

```python
ng.catplot(x='day',y='total_bill', kind='box', data=tip)
```

# regression plot

```python
ng.regplot(x='total_bill', y='tip',data=tip)
```
